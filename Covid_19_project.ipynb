```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df= pd.read_csv("https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv")
df.describe
df.head()
```

```python
#Filtering data for a chosen county
df2=df[df['county']=='New York City']
df2

```

```python
#daily count of cases and deaths
df2.groupby(['cases', 'deaths'])['date'].last()


```

```python
#NYC covid data in the last 30 days
df2.drop(['fips','state'],axis=1).tail(30)


```

```python
#plotting NYC data: rise in cases and deaths in the last few months
df2.plot(kind="scatter", x="cases", y="deaths", color="orange")
```
